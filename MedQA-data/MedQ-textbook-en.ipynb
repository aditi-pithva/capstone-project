### Cleaning the text data, converting to lower case and removing punctuation, extra space

```python
import os
import re

# Path to the folder containing the text files
folder_path = r'C:\Users\daivi\Desktop\Lambton\TERM2\GPT PROJECT\MedQ_Data\textbooks\en'

# Function to clean the text
def clean_text(text):
    # Lowercase the text
    text = text.lower()
    
    # Remove special characters, numbers, and extra spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Keeps only letters and spaces
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    
    return text

# Initialize a dictionary to store cleaned text for each file
cleaned_text_data = {}

# Loop through all files in the folder
for file_name in os.listdir(folder_path):
    # Check if the file is a text file
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)

        # Open and read the text file
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

        # Clean the text
        cleaned_content = clean_text(content)

        # Store the cleaned text in the dictionary with the file name as the key
        cleaned_text_data[file_name] = cleaned_content

# Example: Print the cleaned content of the first file
for file_name, cleaned_text in cleaned_text_data.items():
    print(f"Cleaned content of {file_name}:")
    print(cleaned_text[:500])  # Print the first 500 characters for a preview
    print("\n" + "="*80 + "\n")
```

### Tokenizing the Text

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

files_list = [
    'Anatomy_Gray.txt', 'Biochemistry_Lippincott.txt', 'Cell_Biology_Alberts.txt',
    'First_Aid_Step1.txt', 'First_Aid_Step2.txt', 'Gynecology_Novak.txt', 
    'Histology_Ross.txt', 'Immunology_Janeway.txt', 'InternalMed_Harrison.txt',
    'Neurology_Adams.txt', 'Obstentrics_Williams.txt', 'Pathology_Robbins.txt',
    'Pathoma_Husain.txt', 'Pediatrics_Nelson.txt', 'Pharmacology_Katzung.txt',
    'Physiology_Levy.txt', 'Psichiatry_DSM-5.txt', 'Surgery_Schwartz.txt'
]


# Function to tokenize the cleaned text
def tokenize_text(text):
    return word_tokenize(text)

# Initialize a dictionary to store tokenized text for each file
tokenized_text_data = {}

# Tokenize the cleaned text
for file_name, cleaned_text in cleaned_text_data.items():
    tokenized_content = tokenize_text(cleaned_text)
    tokenized_text_data[file_name] = tokenized_content

# Example: Print tokenized content of one file
print(f"Tokenized content of {files_list[0]}:")
print(tokenized_text_data[files_list[0]][:50])  # Print the first 50 tokens
```
